```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\DELL\OneDrive\Desktop\data_analytics\fortune_1000_companies\fortune_companies.csv')

df.head()

df.describe()

df.shape

df.info
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Sector', data=df, order=df['Sector'].value_counts().index)
plt.title('Distribution of Sectors')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
